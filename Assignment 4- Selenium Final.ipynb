```python
import nltk
from nltk.tokenize import word_tokenize
Match_title=[]
Series=[]
place=[]
Date_of_event=[]
Event_time=[]
Date=[]
try:
    for i in driver.find_elements_by_xpath('//strong[@class="fixture__name fixture__name--with-margin"]'):
        Match_title.append(i.text)
except NoSuchElementException:
    Match_title.append('Unknown')
try:
    for i in driver.find_elements_by_xpath('//span[@class="u-unskewed-text fixture__tournament-label u-truncated"]'):
        Series.append(i.text)
except NoSuchElementException:
    Series.append('Unknown')
try:
    for i in driver.find_elements_by_xpath('/html/body/div[4]/div/div/div[2]/section/div/div/a/div/div/p/span'):
        place.append(i.text)
except NoSuchElementException:
    place.append('Unknown')
try:
    for i in driver.find_elements_by_xpath('//div[@class="fixture__datetime desktop-only"]'):
        Date.append(i.text)
except NoSuchElementException:
    Date.append('Unknown')
for i in range(0,11):

    dt=word_tokenize(Date[i])
    list1=[dt[1],dt[2],'2021']
    list2=[dt[3],dt[4]]
    s="-"
    Date_of_event.append(s.join(list1))
    k=" "
    Event_time.append(k.join(list2))
df2=pd.DataFrame()
df2['Match_title']=Match_title
df2['Series']=Series
df2['place']=place
df2['Date_of_event']=Date_of_event
df2['Event_time']=Event_time
df2.head()
```
