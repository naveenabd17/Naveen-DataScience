```python
# Importing Libraries
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup

# Importing selenium webdriver 
from selenium import webdriver

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException,ElementNotInteractableException

#Importing requests
import requests

# importing regex
import re
```

```python
#calling the driver
driver=webdriver.Chrome('chromedriver.exe')
```

```python
driver.get('https://www.quikr.com/cars/used+cars+all-india+z1399')
```

```python
# to get the url of every car we need xpath and this xpath follows certain pattern so creating the xpath.
sample1=[]
for i in range(1,1000):
    k1='/html/body/div[1]/div/div/div[2]/div/div/div/div/div[1]/div[1]/div[1]/div/div[1]/div/div/div['
    k2=str(i)
    k3=']/a'
    k=k1+k2+k3
    sample1.append(k)
```

```python
car_urls=[]
for i in sample1:
    try:
        car=driver.find_elements_by_xpath(i)
        for i in car:
            car_urls.append(i.get_attribute('href'))
    except NoSuchElementException:
        car_urls.append('No details available')
    except StaleElementReferenceException:
        car_urls.append('No details available')
```

```python
len(car_urls)
```

```python
#Collecting the total information about a car
car_price=[]
car_name=[]
city_name=[]
fuel_type=[]
kil_d=[]
owner_d=[]
year_d=[]
x=0
import time
for j in car_urls:
    print(x)
    driver.get(j)
    time.sleep(2)
    
    #fetching the car price
    try:
        car_p=driver.find_elements_by_xpath('//*[@id="stickyBtnsLeft"]/div[1]/div[1]/label')
        for i in car_p:
            car_price.append(i.text)
    except NoSuchElementException:
        car_price.append('No details available')
    except StaleElementReferenceException:
        car_price.append('No details available')
        
    #fetching the car name
    try:       
        car_1=driver.find_elements_by_xpath('//*[@id="stickyBtnsLeft"]/h1')
        for i in car_1:
            car_name.append(i.text)
    except NoSuchElementException:
        car_name.append('No details available')
    except StaleElementReferenceException:
        car_name.append('No details available')
        
    # fetching the city name
    try:        
        city=driver.find_elements_by_xpath('//*[@id="root"]/div[1]/header/div/div/div[1]/div/label')
        for i in city:
            city_name.append(i.text)
    except NoSuchElementException:
        city_name.append('No details available')
    except StaleElementReferenceException:
        city_name.append('No details available')  
        
    # fetching the fuel type
    try:
        fuel=driver.find_elements_by_xpath('//*[@id="Ad_Details"]/ul/li[6]/span')
        for i in fuel:
            fuel_type.append(i.text)
    except NoSuchElementException:
        fuel_type.append('No details available')
    except StaleElementReferenceException:
        fuel_type.append('No details available')
        
    # fetching the kilometer ran
    try:    
        kil=driver.find_elements_by_xpath('//*[@id="Ad_Details"]/ul/li[5]/span')
        for i in kil:
            kil_d.append(i.text)
    except NoSuchElementException:
        kil_d.append('No details available')
    except StaleElementReferenceException:
        kil_d.append('No details available')
        
    #fetching the type of owner
    try:   
        owner=driver.find_elements_by_xpath('//*[@id="stickyBtnsLeft"]/ul[1]/li[3]')
        for i in owner:
            owner_d.append(i.text)
    except NoSuchElementException:
        owner_d.append('No details available')
    except StaleElementReferenceException:
        owner_d.append('No details available') 
        
    #fetching the year of manufacture
    try:
        year=driver.find_elements_by_xpath('//*[@id="Ad_Details"]/ul/li[4]/span')
        for i in year:
            year_d.append(i.text)
    except NoSuchElementException:
        year_d.append('No details available')
    except StaleElementReferenceException:
        year_d.append('No details available') 
        
    x=x+1
```

```python
import pandas as pd
df=pd.DataFrame()
df['Car Price']=car_price[0:191]
df['Car Name']=car_name[0:191]
df['City Name']=city_name[0:191]
df['Fuel type']=fuel_type[0:191]
df['Kilometer ran']=kil_d[0:191]
df['Year of manufacture']=year_d[0:191]
```
