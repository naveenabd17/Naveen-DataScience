```python
# Importing Libraries
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup

# Importing selenium webdriver 
from selenium import webdriver

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException,ElementNotInteractableException

#Importing requests
import requests

# importing regex
import re
```

```python
url="https://droom.in/cars/used"
```

```python
driver=webdriver.Chrome('chromedriver.exe')
```

```python
driver.get(url)
```

```python
# to get the url of every car we need xpath and this xpath follows certain pattern so creating the xpath.
sample1=[]
for i in range(3,1000):
    k1='//*[@id="root"]/div/div[2]/div/div[3]/div/div[2]/div/div['
    k2=str(i)
    k3=']/div/div/figure/a'
    k=k1+k2+k3
    sample1.append(k)
```

```python
car_urls=[]
for i in sample1:
    try:
        car=driver.find_elements_by_xpath(i)
        for i in car:
            car_urls.append(i.get_attribute('href'))
    except NoSuchElementException:
        car_urls.append('No details available')
    except StaleElementReferenceException:
        car_urls.append('No details available')
```

```python
len(car_urls)
```

```python
#Collecting the total information about a car
car_price=[]
car_name=[]
city_name=[]
fuel_type=[]
kil_d=[]
owner_d=[]
year_d=[]
x=0
import time
for j in car_urls:
    print(x)
    driver.get(j)
    time.sleep(2)
    
    #fetching the car price
    try:
        car_p=driver.find_elements_by_xpath('/html/body/div[8]/div[2]/div[1]/div[2]/div[1]/div[1]/div[5]/div[1]/div[2]')
        for i in car_p:
            car_price.append(i.text)
    except NoSuchElementException:
        car_price.append('No details available')
    except StaleElementReferenceException:
        car_price.append('No details available')
        
    #fetching the car name
    try:       
        car_1=driver.find_elements_by_xpath('/html/body/div[8]/div[2]/div[1]/div[2]/div[1]/div[1]/h1')
        for i in car_1:
            car_name.append(i.text)
    except NoSuchElementException:
        car_name.append('No details available')
    except StaleElementReferenceException:
        car_name.append('No details available')
        
    # fetching the city name
    try:        
        city=driver.find_elements_by_xpath('/html/body/div[8]/div[2]/div[1]/div[1]/div[2]/div/div/ul/li[2]/div')
        for i in city:
            city_name.append(i.text)
    except NoSuchElementException:
        city_name.append('No details available')
    except StaleElementReferenceException:
        city_name.append('No details available')  
        
    # fetching the fuel type
    try:
        fuel=driver.find_elements_by_xpath('/html/body/div[8]/div[2]/div[1]/div[1]/div[2]/div/div/ul/li[6]/div')
        for i in fuel:
            fuel_type.append(i.text)
    except NoSuchElementException:
        fuel_type.append('No details available')
    except StaleElementReferenceException:
        fuel_type.append('No details available')
        
    # fetching the kilometer ran
    try:    
        kil=driver.find_elements_by_xpath('/html/body/div[8]/div[2]/div[1]/div[1]/div[2]/div/div/ul/li[1]/div')
        for i in kil:
            kil_d.append(i.text)
    except NoSuchElementException:
        kil_d.append('No details available')
    except StaleElementReferenceException:
        kil_d.append('No details available')
        
    #fetching the type of owner
    try:   
        owner=driver.find_elements_by_xpath('/html/body/div[8]/div[2]/div[1]/div[1]/div[2]/div/div/ul/li[3]/div')
        for i in owner:
            owner_d.append(i.text)
    except NoSuchElementException:
        owner_d.append('No details available')
    except StaleElementReferenceException:
        owner_d.append('No details available') 
        
    #fetching the year of manufacture
    try:
        year=driver.find_elements_by_xpath('/html/body/div[8]/div[1]/div[1]/ol/li[5]/span')
        for i in year:
            year_d.append(i.text)
    except NoSuchElementException:
        year_d.append('No details available')
    except StaleElementReferenceException:
        year_d.append('No details available') 
        
    x=x+1
```

```python
import pandas as pd
df=pd.DataFrame()
df['Car Price']=car_price[0:406]
df['Car Name']=car_name[0:406]
df['City Name']=city_name[0:406]
df['Fuel type']=fuel_type[0:406]
df['Kilometer ran']=kil_d[0:406]
df['Owner type']=owner_d[0:406]
df['Year of manufacture']=year_d[0:406]
```
